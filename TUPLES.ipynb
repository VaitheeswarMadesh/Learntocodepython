Tuple is one of the Data Structure which can store multiple elements, just as same as lists. Tuple has few key differences when compared to list. 
1. Tuple is represented in ( ) braces.
2. Once you create a tuple it cannot be changed.
3. Tuples are indexed

```python
coordinates = ('5161.8238', '3662381')   #Creating a tuple with ( )
print(coordinates)
print(coordinates[1])
```
