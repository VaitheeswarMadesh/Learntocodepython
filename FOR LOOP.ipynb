SPECIAL TYPE OF LOOP WHICH ALLOWS US TO ITERATE IN THE COLLECTIONS

```python
X= ["APPLE", "BANANA", "MANGO", "GRAPES"]
for X in X:     #FOR EVERY ELEMENT X IN THE LIST X
  print(X)
```

```python
y = "Hafeezuddin"
for y in y:
  print(y)
```

```python
e = [1,2,3,4,5,6,6]
for j in e:
  print(j)
```

```python
for index in range(10):
  print(index)
```

```python
for index in range(10,19):
  print(index)
```

```python
X= ["APPLE", "BANANA", "MANGO", "GRAPES"]
for p in range(len(X)):
  print(p)
```

```python
for index in range(10):
  if index == 0:
    print("This is first Iteration")
  elif index==4:
    print("This is fifth iteration")
  print(index)
```
