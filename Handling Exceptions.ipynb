Lot of programs while running may through you a lot of errors and exceptions that may stop execution of the program. Io counter that try and except can be used to handle such exceptions.

```python
try:
  num = int(input("Enter the Number"))
  print(num)
except:
  print("Invalid Input")

```

CATCHING SPECIFIC ERROR LIKE ZeroDivisionError, Value

```python
try:
  value = 10/0 #exit to except
  num = int(input("Enter the Number"))
  print(num)
except ZeroDivisionError:
  print("DIVISION NOT POSSIBLE")

except ValueError:
  print("Invalid  Input")
```

WE CAN STORE THE ERROR INTO A VARIABLE AND PRINT IT ASWELL

```python
try:
  value = 10/0 #exit to except
  num = int(input("Enter the Number"))
  print(num)
except ZeroDivisionError as err:
  print("DIVISION NOT POSSIBLE")
  print(err) #This prints Division by zero
except ValueError:
  print("Invalid  Input")
```
