Lots of time you deal with large amounts of sequential data. You want to make sure to manage it properly. In python List helps us to organize the data properly & use throughout the program. List uses [   ] squre brackets to create a list. When ever a python intrepretor encounters a List [ ] brackets it assumes it to a LIST

```
fruits = ["Apple", "Pineapple", "Orange", "Banana"]  #This is a List

print(fruits)    #PRINTING OF THE LIST

print(fruits[0])          #PRINING AFTER SLICING THE ELEMENTS IN THE LIST
print(fruits[0:2])
print(fruits[::-1])
print(fruits[:1:-1])
print(fruits[:0:-1])
print(fruits[0:4:2])
print(fruits[-1])
print(fruits[1:])
print(fruits[:1])

fruits = " ".join(fruits)   #removing the barrriers of comma's in between the elements in the list
print(fruits)

fruits = fruits.split(" ")
print(fruits)

#OVER-RIDING THE ELEMENTS OF LIST
fruits[1] = "Strawberry"
print(fruits)

```

A LIST IS REALLY AN AWESOME WAY TO STORE LIST
IT Can store thousands of millions of elements in the list fluently.

```
Numbers = [4,5,5,6,6,7,8,8]
Names = ["James", "Jordan", "Mike", "Alica", "John"]

print(len(Names))   #Here we are printing the lenght (no. of elements in the list)
print(max(Numbers))   #Here we are printing the Maximum value among the elements of the list
```

FUNCTIONS REALLY ALLOW US TO DO SOME COOL STUFF WITH THE LIST LIKE ADDING, APPENDING, EXTENDING AND ALTERING THE LIST

---



```
#EXTEND FUNCTION
Numbers = [4,5,2,657, 5,6,6,7,8,8]
Names = ["James", "Jordan", "Mike", "Alica", "John", "Mike" ]

Names.extend(Numbers)
print(Names)


#APPEND FUNCTION
Names.append("Boyance")   #Appending an string with quotes
print(Names)


Names.append(6)   #Appending an Integer value without quotes
print(Names)

#REMOVE FUNCTION
Names.remove("John")
print(Names)

#INSERT FUCNTION 0 [ TWO PARAMETERS - 1ST IS THE INDEX VALUE WHERE YOU WANT TO INSERT THE ELEMENT & SECOND IS THE ACTUAL ELEMENT.]
Names.insert(1, "Andriona")
print(Names)


#pop() functions basically removes the last element of the list

Names.pop()
print(Names)

#find elements in the list
print(Names.index("Andriona"))


#count the number of similar elements in list
print(Names.count("Mike"))

#Sorting
Numbers.sort() #acsending order
print(Numbers)

#reverse
Numbers.reverse() #descdending order
print(Numbers)

#Coping of list
clonedlist = Numbers.copy()
print(clonedlist)

#TO CLEAR LIST
Names.clear() #This will basically clears the entire elements in the list 
```
