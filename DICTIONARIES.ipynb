Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and does not allow duplicates.

As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

Dictionaries are written with curly brackets, and have keys and values.

Imagine a realworld dictionary having a WORD[KEY]: DEFINITION[VALUE] 



```python
monthConversions = {
    "JAN":"January",
    "Feb":"February",        #Key has to unique    #values need not be unique
    "Mar":"March",
}

#Multiple ways for retriveing the values using keys from dictionaries
print(monthConversions)

print(monthConversions.get("JAN"))

print(monthConversions.get("k")) # AS NO SUCH KEY IS EXISTING THE FUNCTION GET RETURNS NONE
```

```python
monthConversions = {
    "JAN":"January",
    "Feb":"February",        #Key has to unique    #values need not be unique
    "Mar":"March",
}

print(monthConversions.get("JAN", "NOT A VALID KEY"))

print(monthConversions.get("k", "NOT A VALID KEY"))
```

```python
monthConversions = {
    "JAN":"January",
    "Feb":"February",        #Key has to unique    #values need not be unique
    "Mar":"March",
}


print(monthConversions)
print(len(monthConversions))
```

```python
thisdict = {
  "brand": "Ford",
  "electric": False,
  "year": 1964,
  "colors": ["red", "white", "blue"]    #can use list as value for a key
}

print(type(thisdict))

print(thisdict["colors"])
```

```python
thisdict = {
  "brand": "Ford",
  "electric": False, #HERE WE HAVE USED BOOLEAN VALUE FOR THE KEY ELECTRIC
  "year": 1964,
}

x = thisdict.keys()  #retriving the key values
print(x)

thisdict["electric"]="true"  #ALTERING THE VALUES
thisdict["OWNER"]="INDIAN"  #ADDING NEW KEY AND VALUE

print(thisdict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"color": "red"})
print(thisdict)
```
