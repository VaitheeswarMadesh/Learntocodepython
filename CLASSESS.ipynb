MAIN TYPES OF DATA WE HAVE SEEN SO FOR STRINGS, NUM, LIST, SETS, TUPLES, DICTIONARIES. 

NOT ALL THINGS CAN FIT INTO ABOVE SPECIFIED DATA TYPES

HERE OBJECTS AND CLASSES COMES TO OUR PICTURE

AN OBJECT IS AN REAL WORLD ENTITY - A PHONE, COMPUTER

WITH CLASSES AND OBJECTS WE CAN CREATE OUR OWN DT AND FIT OUR DATA INTO IT

CLASSES CONTAIN THE ATRIBUTES

```python
class student:   #its basically a student data type 
    def__init__(self, name,major, gpa, is_on_probation):
      self.name = name     #attributes
      self.major = major
      self.gpa = gpa
      self.is_on_probation = is_on_probation

#student is a object here after

student1 = student("jim", "business", "3.1", "False") 
student2 = student("Pam", "business", "3.2", "False") 

print(student1)
print(student2.gpa)

```
